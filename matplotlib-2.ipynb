## Matplotlib Tutorial-2



```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#drawing a pie chart
exp_labels= ['food', 'mobile', 'books', 'movies', 'travel']
exp_vals=[2000,500,200,400,750]
```

```python
plt.pie(exp_vals, labels=exp_labels)
plt.show()
```

```python
#making the axis equal
plt.axis("equal")
plt.pie(exp_vals, labels=exp_labels)
plt.show()
```

```python
#making the bigger radius
plt.axis("equal")
plt.pie(exp_vals, labels=exp_labels,radius=2)
plt.show()
```

```python
#making the  percentage show
plt.axis("equal")
plt.pie(exp_vals, labels=exp_labels,autopct="%0.2f%%")
plt.show()
```

```python
#making the  percentage show
plt.axis("equal")
plt.pie(exp_vals, labels=exp_labels,autopct="%0.1f%%")
plt.show()
```

```python
#making the  pie obvious - using explode
plt.axis("equal")
plt.pie(exp_vals, labels=exp_labels, autopct="%0.0f%%", explode=[0,0,0.2,0,0])
plt.show()
```

```python
#making the  pie obvious - using explode
plt.axis("equal")
plt.pie(exp_vals, labels=exp_labels, autopct="%0.0f%%", explode=[0,0,0.2,0,0], shadow=True)
plt.show()
```

### Drawing Histograms

```python
#drawing histograms

marks = [45, 67,89, 90, 34, 88, 56, 99, 34, 67, 87]
```

```python
plt.hist(marks)
plt.show()
#default is 10 bins
```

```python
#make bin size 3
plt.hist(marks, bins=3)
plt.show()
```

```python
#make bin size 3  - add relative width to get clear picture
plt.hist(marks, bins=3, rwidth =0.95)
plt.show()
```

```python
#make bin size 3
plt.hist(marks, bins=3, rwidth =0.95, cumulative=True)
plt.show()
```
